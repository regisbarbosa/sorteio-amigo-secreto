### Sorteio Amigo Secreto 

```python
# Sorteio de Amigo Secreto com envio de e-mails e restrições
import random
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

# Função para validar e-mails
def validar_emails(nomes_emails):
    for nome, email in nomes_emails.items():
        if "@" not in email or "." not in email:
            raise ValueError(f"E-mail inválido para {nome}: {email}")

# Função principal de sorteio
def sortear_amigo_secreto(nomes, restricoes):
    tentativas = 0
    while True:
        tentativas += 1
        random.shuffle(nomes)
        resultado = {nomes[i]: nomes[(i + 1) % len(nomes)] for i in range(len(nomes))}

        # Verificar autoatribuição e restrições
        if all(
            sorteado != amigo and sorteado not in restricoes.get(amigo, [])
            for amigo, sorteado in resultado.items()
        ):
            print(f"Sorteio realizado após {tentativas} tentativa(s).")
            return resultado

# Função para enviar e-mails
def enviar_email(amigo, sorteado, email_amigo, email_remetente, senha_remetente):
    try:
        msg = MIMEMultipart()
        msg["From"] = email_remetente
        msg["To"] = email_amigo
        msg["Subject"] = "Seu Amigo Secreto!"

        corpo = f"Olá, {amigo}!\n\nVocê tirou {sorteado} no amigo secreto!\n\nDivirta-se!"
        msg.attach(MIMEText(corpo, "plain"))

        with smtplib.SMTP("smtp.gmail.com", 587) as server:
            server.starttls()
            server.login(email_remetente, senha_remetente)
            server.sendmail(email_remetente, email_amigo, msg.as_string())
        print(f"E-mail enviado para {amigo} ({email_amigo}).")
    except Exception as e:
        print(f"Erro ao enviar e-mail para {amigo}: {e}")

# Configurações iniciais
nomes_emails = {
    "Alice": "alice@gmail.com",
    "Bob": "bob@gmail.com",
    "Carol": "carol@gmail.com"
}

restricoes = {
    "Alice": ["Bob"],
    "Bob": ["Alice"]
}

email_remetente = "seu_email@gmail.com"
senha_remetente = "sua_senha"

# Validação de e-mails
try:
    validar_emails(nomes_emails)
except ValueError as e:
    print(e)
    exit()

# Execução do sorteio e envio de e-mails
try:
    resultado = sortear_amigo_secreto(list(nomes_emails.keys()), restricoes)
    for amigo, sorteado in resultado.items():
        enviar_email(amigo, sorteado, nomes_emails[amigo], email_remetente, senha_remetente)
except Exception as e:
    print(f"Erro durante o sorteio ou envio de e-mails: {e}")
```
